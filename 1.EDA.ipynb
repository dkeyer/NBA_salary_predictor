Find players from all 5 years, output their averages

```python
s=pd.concat([twenty_stats,nineteen_stats,eighteen_stats,seventeen_stats,sixteen_stats],axis=0)
```

Above, I am making a dataframe of the players that have been in the league for at least 5 years 
(so each will be in the last 5 years of dataframes). I will then add the players' salaries so I can
see how their extended performances (not just within a year) affect their salaries
This keeps me from having to drop rookies individually, as they will not meet the 5 year criteria to be in this dataset,
and so will be dropped. The number of years is subject to change, and I am only using 5 as a starting point
Furthermore, I will use the same method to predict for slaries before 2020 using past statistics

```python
new_s = s[s['Player'].map(s['Player'].value_counts()) == 5]
new_s

#Here, we are grouping by players who are in all five of our datasets
#The next step will be to take their averages from the five seasons
```

```python
aggregation_functions = {'Rk': 'first', 'Season': 'first', 'Pos': 'first', 'Age': 'mean', 'Tm': 'first', 'G': 'mean', 'GS': 'mean', 'MP': 'mean', 'FG': 'mean',
       'FGA': 'mean', 'FG%': 'mean', '3P': 'mean', '3PA': 'mean', '3P%': 'mean', '2P': 'mean', '2PA': 'mean', '2P%': 'mean', 'eFG%': 'mean', 
        'FT': 'mean', 'FTA': 'mean', 'FT%': 'mean', 'ORB': 'mean', 'DRB': 'mean', 'TRB': 'mean', 'AST': 'mean', 'STL': 'mean',
        'BLK': 'mean', 'TOV': 'mean', 'PF': 'mean',
       'PTS': 'mean'}
df_new = new_s.groupby(new_s['Player']).aggregate(aggregation_functions)
df_new
#Finding averages of all of the players over this five year span
```

```python
sixteen_to_twenty_with_sal = pd.merge(left=df_new, right=twenty_salary_cond, how='left', on='Player')
#Adding 19-20 salary to dataframe
```

```python
sixteen_to_twenty_with_sal['Player'].duplicated().value_counts()
#We have some duplicates we need to drop
```

A quick aside: When I work with data from the other years (roll stats from multiple years into one dataframe) I will use the same method as I used above. Those years are not included here because I am not yet exactly sure which they will be.
